```python
import pandas as pd
import matplotlib.pyplot as plt

def channel_breakout(df, window=20, n_std=2):
    rolling_mean = df['Close'].rolling(window=window).mean()
    rolling_std = df['Close'].rolling(window=window).std()
    upper_band = rolling_mean + (rolling_std * n_std)
    lower_band = rolling_mean - (rolling_std * n_std)

    buy_signal = None
    sell_signal = None
    hold_signal=None
    for i in range(window, len(df)):
        if df.iloc[i]['Close'] > upper_band[i]:
            sell_signal = i
        elif df.iloc[i]['Close'] < lower_band[i]:
            buy_signal = i
        else:
            hold_signal=i
    return buy_signal, sell_signal,hold_signal, rolling_mean, upper_band, lower_band
```

```python
class TrafficVolumeDataset(torch.utils.data.Dataset):
  def __init__(self, train=False, validate=False, test=False, window_size=500):

    #STEP1: Load the data
    self.df_train = df_train
    self.df_val = df_val
    self.df_test = df_test

    #STEP2: Creating Features
    if train: #process train dataset
      features = self.df_train[['Open', 'High', 'Low', 'Close']]
      target = self.df_train.Position
    elif validate: #process validate dataset
      features = self.df_val[['Open', 'High', 'Low', 'Close']]
      target = self.df_val.Position
    else: #process test dataset
      features = self.df_test[['Open', 'High', 'Low', 'Close']]
      target = self.df_test.Position

    #STEP3: Create windows/sequencing
    self.x, self.y = [], []
    for i in range(len(features) - window_size):
        v = features.iloc[i:(i + window_size)].values
        self.x.append(v)
        self.y.append(target.iloc[i + window_size])

    #STEP4: Calculate length of dataset
    self.num_sample = len(self.x)

  def __getitem__(self, index):
    x = self.x[index].astype(np.float32)
    y = self.y[index].astype(np.float32)
    return x, y

  def __len__(self):
    #returns the total number of records for data set
    return self.num_sample
```

```python
class TrafficVolumePrediction(pl.LightningModule):
  def __init__(self, input_size=4, output_size=1, hidden_dim=10, n_layers=2,
               window_size=500):
    """
    input_size: Number of features in the input
    hidden_dim: number of hidden layers
    n_layers: number of RNN to stack over each other
    output_size: number of items to be outputted
    """
    super(TrafficVolumePrediction, self).__init__()

    self.hidden_dim = hidden_dim
    self.n_layers = n_layers
    self.TrafficVolumePrediction = nn.LSTM(input_size, hidden_dim, n_layers,
                                           bidirectional=False, batch_first=True)
    self.fc = nn.Linear(hidden_dim * window_size, output_size)
    self.loss = nn.BCELoss()
    self.learning_rate = 0.001

    # self.relu = nn.ReLU()

  def forward(self, x):
    batch_size = x.size(0)
    hidden = self.get_hidden(batch_size)
    out, hidden = self.TrafficVolumePrediction(x, hidden)
    # out = self.relu(out)
    out =out.reshape(out.shape[0], -1)
    out = self.fc(out)
    out = torch.sigmoid(out)
    return out


  def get_hidden(self, batch_size):
    # hidden = torch.zeros(self.n_layers, batch_size, self.hidden_dim)
    hidden_state = torch.zeros(self.n_layers, batch_size, self.hidden_dim)
    cell_state = torch.zeros(self.n_layers, batch_size, self.hidden_dim)
    hidden = (hidden_state, cell_state)
    return hidden

  def configure_optimizers(self):
    params = self.parameters()
    optimizer = optim.Adam(params=params, lr = self.learning_rate)
    # optimizer = optim.RMSprop(params=params, lr = 0.001)
    return optimizer

  def training_step(self, train_batch, batch_idx):
    features, targets = train_batch
    output = self(features)
    output = output.view(-1)
    loss = self.loss(output, targets)
    self.log('train_loss', loss, prog_bar=True)
    return {"loss": loss}

  def train_dataloader(self):
    traffic_volume_train =  TrafficVolumeDataset(train=True)
    train_dataloader = torch.utils.data.DataLoader(traffic_volume_train, batch_size=50)
    return train_dataloader

  def validation_step(self, val_batch, batch_idx):
    features, targets = val_batch
    output = self(features)
    output = output.view(-1)
    loss = self.loss(output, targets)
    self.log('val_loss', loss, prog_bar=True)

  def val_dataloader(self):
    traffic_volume_val =  TrafficVolumeDataset(validate=True)
    val_dataloader = torch.utils.data.DataLoader(traffic_volume_val, batch_size=50)
    return val_dataloader
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Получение данных
# data = pd.read_csv('path/to/data.csv', index_col=0)
data = df.copy()
# Построение временного ряда
prices = data['Low']
plt.plot(prices)

# Выбор типа скользящих средних
sma_prices = prices.rolling(window=30).mean() # простое скользящее среднее
wma_prices = prices.rolling(window=30, win_type='triang').mean() # взвешенное скользящее среднее
ema_prices = prices.ewm(span=30, adjust=False).mean() # экспоненциальное скользящее среднее

# Подбор оптимального периода
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error

splits = TimeSeriesSplit(n_splits=5)
best_rmse = float('inf')
best_period = None
best_type = None

for period in range(10, 100):
    for type in ['sma', 'wma', 'ema']:
        if type == 'sma':
            prices = sma_prices
        elif type == 'wma':
            prices = wma_prices
        elif type == 'ema':
            prices = ema_prices

        rmse_scores = []
        for train_index, test_index in splits.split(prices):
            train, test = prices[train_index], prices[test_index]
            predictions = pd.Series([train[-1]] * len(test), index=test.index)
            for i in range(len(test)):
                if test.iloc[i] > predictions.iloc[i]:
                    predictions.iloc[i] = train.mean()
                else:
                    predictions.iloc[i] = train.min()
            rmse_scores.append(mean_squared_error(test, predictions, squared=False))
        avg_rmse = sum(rmse_scores) / len(rmse_scores)
        if avg_rmse < best_rmse:
            best_rmse = avg_rmse
            best_period = period
            best_type = type

# Применение скользящих средних
if best_type == 'sma':
    prices = sma_prices
elif best_type == 'wma':
    prices = wma_prices
elif best_type == 'ema':
    prices = ema_prices

predictions = pd.Series([prices.iloc[0]] * len(prices), index=prices.index)
for i in range(1, len(prices)):
    if prices.iloc[i] > predictions.iloc[i-1]:
        predictions.iloc[i] = prices.iloc[i-best_period:i].mean()
    else:
        predictions.iloc[i] = prices.iloc[i-best_period:i].min()

# Оценка результатов
plt.plot(predictions)
plt.legend(['Actual', 'Predicted'])
plt.show()
print(f'Best type: {best_type}, Best period: {best_period}')
```

Затем мы выбираем три типа скользящих средних (SMA, WMA и EMA) и вычисляем их значения с помощью метода `rolling` для SMA и WMA и метода `ewm` для EMA.

Затем мы подбираем оптимальный период и тип скользящих средних с помощью метода перекрестной проверки и выбираем тип и период с наименьшим значением среднеквадратической ошибки (RMSE).
